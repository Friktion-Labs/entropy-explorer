# ü•≠ Show Perp Fills

This notebook shows how to fetch the event queue for a perp market, and list all the fills.

```python
import entropy

with entropy.ContextBuilder.build(cluster_name="mainnet") as context:
    market = entropy.market(context, "BTC-PERP")
    event_queue = entropy.PerpEventQueue.load(context, market.event_queue_address, market.lot_size_converter)
    print(event_queue.fills)

print("Example complete.")
```
